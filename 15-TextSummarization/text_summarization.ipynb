```python
chat_message = [
    SystemMessage(content='You ar an expert with expertise in summarizing speeches'),
    HumanMessage(content=f'Please provide a concise summary of the following speech.\nText:{speech}')
]
```

```python
from langchain_groq import ChatGroq
import os
from dotenv import load_dotenv
load_dotenv()

llm = ChatGroq(model="gemma2-9b-it")
llm
```

## Prompt template text summarization

```python
from langchain.chains.llm import LLMChain
```

```python
LLMChain
```

```python
from langchain.prompts import PromptTemplate
generic_prompt_template = '''
Write a concise summary of the following speech:
Speech : {speech}
Translate the summary to {language} language
'''
prompt = PromptTemplate.from_template(
    template=generic_prompt_template
    )
prompt
```

```python
final_prompt = prompt.format(language='Hindi',speech=speech)
final_prompt
```

```python
llm.get_num_tokens(final_prompt)
```

```python
llm_chain = LLMChain(llm=llm,prompt=prompt)
llm_chain
```

```python
summary = llm_chain.run({"speech": speech,"language":"Hindi"})
summary
```

### Summarization using Map Reduce

```python
docs 
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
splitter = RecursiveCharacterTextSplitter(chunk_size=2500,chunk_overlap=50)
split_docs = splitter.split_documents(docs)
len(split_docs)
```

```python
prompt
```

```python
chunk_prompt = '''
Summarize the below speech:
{text}
Summary:
'''
map_prompt_template = PromptTemplate.from_template(chunk_prompt)
map_prompt_template
```

```python
final_prompt = ''' Provide a final summary of the entire speech based on following points.
Add a motivational title,Start the concise summary with an introduction and provide the summary in number points for the speech below
Speech : {text}
'''
final_prompt_template = PromptTemplate.from_template(final_prompt)
final_prompt_template
```

```python
summary_chain = load_summarize_chain(
    llm=llm,
    chain_type="map_reduce",
    map_prompt = map_prompt_template,
    verbose=True,
    combine_prompt = final_prompt_template
    )
summary_chain
```

```python
output = summary_chain.run(split_docs)
```

```python
output
```
